```python
import pickle
# Load the dataset using the uploaded file path
df = pd.read_csv("Student Mental health.csv")

# Rename columns for better readability
df.columns = ['Time', 'Gender', 'Age', 'Course', 'Year', 'CGPA', 'Marriage', 'Depression', 'Anxiety', 'Panic', 'Treatment']

# Handle missing values
df.dropna(inplace=True)

# Extract year information
new = df['Year'].str.split(' ', n=1, expand=True)
df['Year'] = new[1]

# Clean CGPA values
df["CGPA"] = df["CGPA"].str.strip(" ")

# Convert 'Age' to integer
df['Age'] = df['Age'].astype(int)

# Drop 'Time' column as it's not needed for prediction
df.drop("Time", axis=1, inplace=True)

# Convert 'Year' to integer
df['Year'] = df['Year'].astype(int)

# Replace CGPA ranges with numeric values
df.replace("0 - 1.99", "0", inplace=True)
df.replace('2.00 - 2.49', "1", inplace=True)
df.replace('2.50 - 2.99', "2", inplace=True)
df.replace('3.50 - 4.00', "3", inplace=True)
df.replace('3.00 - 3.49', "4", inplace=True)
df['CGPA'] = df['CGPA'].astype('category')

# Encode categorical values as numeric
df['Gender'] = df['Gender'].apply(lambda x: 0 if x == 'Female' else 1).astype('category')
for column in ['Marriage', 'Depression', 'Anxiety', 'Panic', 'Treatment']:
    df[column] = df[column].apply(lambda x: 0 if x == 'No' else 1).astype('category')

# Convert categorical 'Course' into dummy variables
df = pd.get_dummies(df, columns=['Course'])

# Prepare training and test sets
X = df.drop(["Depression"], axis=1)
y = df["Depression"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

# Initialize and train the RandomForest model
RF = RandomForestClassifier(
    max_depth=13,
    max_features='log2',
    min_samples_leaf=1,
    min_samples_split=5,
    n_estimators=500
)
RF.fit(X_train, y_train)

# Evaluate model performance
ypred = RF.predict(X_test)
cm = confusion_matrix(y_test, ypred)
accuracy = accuracy_score(y_test, ypred)
f1_macro = f1_score(y_test, ypred, average='macro')

# Save the trained model as a pickle file
with open('RandomForestModel.pkl', 'wb') as file:
    pickle.dump(RF, file)

# Provide the path, confusion matrix, accuracy, and F1 score
("RandomForestModel.pkl", cm, accuracy, f1_macro)
```
